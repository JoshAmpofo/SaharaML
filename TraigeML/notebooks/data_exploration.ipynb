```python
# check columns in both datasets
print(f"Columns in DiseaseAndSymptoms Dataset:\n{data.columns}\n\nColumns in Disease Precautions Dataset:\n{precaus.columns}")
```

```python
# check diseases in both datasets
print(f"Disease count in disease dataset: {data['Disease'].nunique()}")
print(f"Disease count in precautions dataset: {precaus['Disease'].nunique()}")
```

```python
# clean up the inconsistent namings
data['Disease'] = data['Disease'].str.replace(" ", "_").str.lower()
precaus['Disease'] = precaus['Disease'].str.replace(" ", "_").str.lower()

# check diseases in both datasets after cleaning
print(f"Diseases in disease dataset after cleaning: {set(data['Disease'].to_list())}")
print(f"Diseases in precautions dataset after cleaning: {set(precaus['Disease'].to_list())}")
```

```python
# check for null values
print(data.isnull().sum(), "\n")
print(precaus.isnull().sum())
```

### **Data Distribution**

```python
# visualize disease count
plt.figure(figsize=(12,8))
sns.countplot(y='Disease', data=data, order=data['Disease'].value_counts().index)
plt.title('Disease Count in DiseaseAndSymptoms Dataset')
plt.show()
```

```python
# disease count in precautions dataset
plt.figure(figsize=(12,8))
sns.countplot(y='Disease', data=precaus, order=precaus['Disease'].value_counts().index)
plt.title('Disease Count in Disease Precautions Dataset')
plt.show()
```

```python
# symptoms
symptom_cols = data.columns[1:]
symptom_cols
data[symptom_cols]
```

```python
data[symptom_cols]['Symptom_1'].value_counts()
```

```python
# plot top 10 symptoms in Symptom_1
plt.figure(figsize=(10,6))
top_10_symptoms = data[symptom_cols]['Symptom_1'].value_counts().head(10)
sns.barplot(x=top_10_symptoms.values, y=top_10_symptoms.index)
plt.title('Top 10 Symptoms in Symptom_1')
plt.xlabel('Count')
plt.ylabel('Symptoms')
plt.show()
```

```python
# create a grid of subplots of the top ten symptoms in each symptom column
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 15))
axes = axes.flatten()
for i, col in enumerate(symptom_cols[:9]):
    top_10_symptoms = data[symptom_cols][col].value_counts().head(10)
    sns.barplot(x=top_10_symptoms.values, y=top_10_symptoms.index, ax=axes[i])
    axes[i].set_title(f'Top 10 Symptoms in {col}')
    axes[i].set_xlabel('Count')
    axes[i].set_ylabel('Symptoms')
plt.tight_layout()
plt.show()
```

```python
# precautions columns
precaus_cols = precaus.columns[1:]
precaus_cols
precaus[precaus_cols]
```

```python
# create a grid of subplots of the top ten precautions in each symptom column
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 15))
axes = axes.flatten()
for i, col in enumerate(precaus_cols[:4]):
    top_10_precautions = precaus[precaus_cols][col].value_counts().head(10)
    sns.barplot(x=top_10_precautions.values, y=top_10_precautions.index, ax=axes[i])
    axes[i].set_title(f'Top 10 Precautions in {col}')
    axes[i].set_xlabel('Count')
    axes[i].set_ylabel('Precautions')
plt.tight_layout()
plt.show()
```
